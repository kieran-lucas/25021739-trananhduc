```python
#Problem 8
def reverse(n):
    s = 0
    while n > 0:
        s = s*10 + n % 10
        n //= 10 
    return s
n = int(input())
s = 0
while n != reverse(n):
    n += reverse(n)
    s += 1
print(n, s)
    

        
    
```

```python
#Problem 10
n = int(input())
def collatz(x):
    count = 1
    while x != 1:
        if x % 2 == 0:
            x //= 2
        else:
            x = 3*x + 1
        count += 1
    return count
best_x = 1
best_len = 1
for i in range(1, n + 1):
    l = collatz(i)
    if l > best_len or (l == best_len and i < best_x):
        best_x = i
        best_len = l
print(best_x, best_len)
            
            
```

```python
#Problem 13
def close_pair(n):
    s = 0
    for i in range(1, n):
        if n % i == 0:
            s += i
    return s
a, b = map(int, input().split())
if close_pair(a) == b and close_pair(b) == a:
    print('True')
else:
    print('False')
    
```
