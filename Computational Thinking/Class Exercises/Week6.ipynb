```python
#W6A1
n = list(map(int, input().split()))
a = set()
b = []
for i in n:
    if i not in a:
        a.add(i)
        b.append(i)
print(b)
```

```python
#W6A2
n = list(map(int, input().split()))
a = []
for i in range(len(n)):
    a.append(sum(n[i] for i in range(i + 1)))
print(a)
```

```python
#W6A4
pairs = input().split()
d = {}
for p in pairs:
    key, value = p.split(":")
    if key not in d:
        d[key] = []
    d[key].append(value)
print(d)
```

```python
#W6A5
n = list(map(int, input().split()))
d = {'positives' : 0, 'negatives' : 0, 'zero' : 0}
for i in n:
    if i > 0:
        d['positives'] += 1
    elif i < 0:
        d['negatives'] += 1
    else:
        d['zero'] += 1
print(d)
        
        
        
        
```

```python
#W6A9
d1 = eval(input())
d2 = eval(input())
result = {}
for k in d1:
    result[k] = d1[k]
for k in d2:
    result[k] = result.get(k, 0) + d2[k]
for k in sorted(result):
    print(k, ":", result[k])
```

```python
#W6A10
my_list = list(map(int, input().split()))
k = int(input())
list_number = []
for i in range(len(my_list) - 1):
    for j in range(i + 1, len(my_list)):
        if my_list[i] + my_list[j] == k:
            list_number.append((my_list[i], my_list[j]))
list_number.sort(key = lambda x: x[0])
print(list_number)
            
            
```

```python
#W6A11
my_tuple = tuple(map(int, input().split()))
even_tuple = []
odd_tuple = []
for x in my_tuple:
    if x % 2 == 0:
        even_tuple.append(x)
    if x % 2 != 0:
        odd_tuple.append(x)
print(tuple(even_tuple), "\n", tuple(odd_tuple))
```

```python
#W6A12
my_list = list(map(int, input().split()))
count = {}
for k in my_list:
    count[k] = count.get(k, 0) + 1
max_count = max(count.values())
number = min(k for k, v in count.items() if count[k] == max_count)
print(number)
```

```python
#W6A15
items = input().split()
k = int(input())
d = {items[i]: items[i + 1] for i in range(0, len(items), 2)}
e = {}
for key, value in d.items():
    if int(value) > k:
        e[key] = int(value)
print(e)
```
