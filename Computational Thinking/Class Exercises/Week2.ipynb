```python
#Problem 1
'''
Input a number and check if it is even or odd
'''

a = int(input("Enter a integer number: "))
match a % 2:
    case 0: 
        print(f"{a} is even")
    case 1:
        print(f"{a} is odd")


```

```python
#Problem 2
'''
Input a letter grade (A, B, C, D, F) and print its evaluation:
"Excellent" → A
"Good" → B
"Average" → C
"Poor" → D
"Fail" → F
'''

a = input("Enter a letter grade: ")
match a.upper():
    case "A":
        print("Excellent")
    case "B":
        print("Good")
    case "C":
        print("Average")
    case "D":
        print("Poor")
    case "F": 
        print("Fail")
```
