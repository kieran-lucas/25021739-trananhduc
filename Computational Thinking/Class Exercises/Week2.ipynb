```python
#Validate the date
#month with 30 days: 4 6 9 11
#month with 31 days: 1 3 5 7 8 10 12
#month with 28, 29 days: 2
#leap year: divisible by 400 or ( divisible by 4 but not divisible by 100)
a,b,c = map(int, input("Enter the date: ").split("/"))
#a/b/c
if (a < 1 or a > 31) or (b < 1 or b > 12) or (a < 1900 or a > 2025):
    print("Invalid")
elif (b == 4 or b == 6 or b == 9 or b == 11) and (a == 31):
    print("Invalid")
elif not (c % 400 == 0 or (c % 4 == 0 and c % 100 != 0)) and (c == 29):
    print("Invalid")
else: 
    print("Valid")
```
