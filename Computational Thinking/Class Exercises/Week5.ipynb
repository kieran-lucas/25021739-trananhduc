```python
#W5A7
def calculator(num1, num2, operator):
    if operator == "+":
        return num1 + num2
    elif operator == "-":
        return num1 - num2
    elif operator == "*":
        return num1*num2
    elif operator == "/":
        return round(num1/num2, 2)
    else:
        return -1
num1, num2 = map(int, input().split())
operator = input()
calculator(num1, num2, operator)
    
```

```python
#W5A8
def hamming_distance(x, y):
    result = x ^ y #Trả về 1 nếu hai bit tương ứng khác nhau
    count = 0
    while result > 0:
        count += result & 1 #Thêm 1 nếu bit bên phải ngoài cùng
        #của biến result là 1 (để đếm số bit 1 của result)
        result >>= 1 #Dịch bit sang phải 1 đơn vị của result
        #để kiểm tra bit tiếp theo cho đến khi result = 0
    return count
    #Biến count được trả về chính là khoảng cách hamming 
    #của x và y
x, y = map(int, input().split())
print(hamming_distance(x, y))
```

```python
#W5A9
def sum_of_digits(n):
    s = 0
    while n > 0:
        s += n % 10
        n //= 10
    return s
n = int(input())
sum_of_digits(n)
```

```python
#W5A10
def iso(a, b):
    for i in range(len(a)):
        for j in range(i + 1, len(a)):
            if (a[i] == a[j] and b[i] != b[j]) or (a[i] != a[j] and b[i] == b[j]):
                return False
    return True
    
a, b = input().split()
if iso(a, b):
    print("Dang cau")
else:
    print("Khong dang cau")
    
```
