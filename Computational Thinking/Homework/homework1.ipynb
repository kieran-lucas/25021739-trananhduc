```python
#problem 1
"""
Nhập vào số nguyên n, in ra màn hình số gấp đôi của n.
"""

a = int(input("Enter a integer number: "))
print(2*a)
```

```python
#problem 2
"""
Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. 
Người ta xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. 
Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. 
Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân.
"""

a, b = map(int, input("Enter a and b: ").split())
c = a*b - 3.14*a*a/4
print(f"The area to plan tree is {c:.2f}")
```

```python
#problem 3
"""
Viết chương trình nhập vào một ký tự c. 
Nếu nó là chữ hoa thì in ra chữ thường tương ứng 
và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. 
"""

a = input("Enter a character: ")
if a.isupper():
    print(a.lower())
elif a.islower():
    print(a.upper())
```

```python
#problem 4
"""
Viết chương trình nhập vào một ký tự c, 
kiểm tra xem c có phải là ký tự chữ cái không (hoa hoặc thường). 
Nếu đúng thì in ra “{c} là ký tự alphabet”, 
nếu sai in ra “{c} không phải là ký tự alphabet.”
"""

a = input("Enter a character: ")
if a.isalpha():
    print(f"{a} is a alphabet character")
else:
    print(f"{a} is not a alphabet character")
    
```

```python
#problem 5
"""
Nhập vào chữ cái hoa A, 
in ra chữ cái thường liền trước chữ cái a thường tương ứng của A. 
Xét trường hợp khi c = ‘A’ là đặc biệt.
"""

a = input("Enter the upper character: ")
if a == "A":
    print("This is a special character")
else: 
    b = chr(ord(a) + 31)
    print(b)
```

```python
#problem 6
"""
Nhập vào 3 số a, b, c. 
Kiểm tra xem đây có phải là 3 cạnh của 1 tam giác không. 
Nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân. 
Nếu không phải thì in ra “Không phải 3 cạnh của tam giác”.
"""


a, b, c = map(float, input("Enter 3 side of triangle: ").split())
if a + b > c and b + c > a and c + a > b:
    p = (a + b + c)/2
    s=(p*(p-a)*(p-b)*(p-c))**0.5
    print(f"This is a triangle, and the area is {s:.2f}")
else: 
    print("This is not a triangle")
```

```python
#prolem 8
"""
Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này, 
tính tiền điện tháng này cho hộ biết:

- Từ 0 – 50 kWh: 1.984 đồng/kWh
- Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh

Tính thêm tiền VAT 8%. Chỉ lấy đến đơn vị đồng.

Ví dụ:
Input:
Ten chu ho: Pham Xuan Toai
Chi so thang truoc: 1020
Chi so thang nay: 1070

Output:
Ho va ten: Pham Xuan Toai
Tien phai tra la: 107136
"""

t = input("Enter the owner's name: ")
b,c = map(int, input("Enter last month, this month index: ").split())
a = c - b
if a <= 50:
    d = 50*1.984
elif a <= 100:
    d = 50*1.984 + (d-50)*2.050
elif a <= 200:
    d = 50*1.984 + d*50 + (d-100)*2.380
elif a <= 300:
    d = 50*1.984 + d*50 + d*2.380 + (d-200)*2.998
elif a <= 400:
    d = 50*1.984 + d*50 + d*2.380 + d*2.998 + (d-300)*3.350
else:
    d = 50*1.984 + d*50 + d*2.380 + d*2.998 + d*3.350 + (d-400)*3.460
total = d*10800
print(f" Ho va ten: {t} \n The bill is: {total:.0f} VND")
```

```python
#problem 1
"""
Kiểm tra số chẵn lẻ: 
Viết chương trình nhập một số nguyên, kiểm tra và in ra số đó là số chẵn hay số lẻ.
"""

a = int(input("Enter a integer number: "))
if a % 2 == 0:
    print(f"{a} is even")
else:
    print(f"{a} is odd")
    
```

```python
#problem 4
"""
Kiểm tra tuổi đủ điều kiện bầu cử: 
Viết chương trình nhập năm sinh của một người, 
kiểm tra xem người đó có đủ 18 tuổi để đi bầu cử hay không.
"""

a = int(input("Enter the birth year: "))
b = 2025 - a
if b >= 18:
    print("This person is old enough to vote")
else: 
    print("This person is not old enough to vote")
    
```

```python
#problem 5
"""
So sánh 2 số: 
Viết chương trình nhập vào hai số, in ra số lớn hơn 
hoặc thông báo nếu hai số bằng nhau.
"""

a, b = map(float, input("Enter two number: ").split())
if a > b:
    print(f"{a:.0f} is larger than {b:.0f}")
elif a < b:
    print(f"{a:.0f} is smaller than {b:.0f}")
else:
    print(f"{a:.0f} is equal to {b:.f0}")
```

```python
#problem 6
"""
Kiểm tra ký tự nhập vào là chữ hay số: 
Viết chương trình nhập một ký tự, kiểm tra xem ký tự đó là chữ cái hay số.
"""

a = input("Enter a character: ")
if a.isalpha():
    print(f"{a} is a alphabet character")
else:
    print(f"{a} is a number")
```

```python
#problem 7
"""
Tính điểm học lực  
Nhập điểm (0–10), phân loại học lực theo quy tắc:  
>=8: Giỏi  
>=6.5 và <8: Khá  
>=5 và <6.5: Trung bình  
<5: Yếu
"""

a = float(input("Enter the score: "))
if a >= 8:
    print("Good")
elif a >= 6.5:
    print("Fair")
elif a >= 5:
    print("Average")
else:
    print("Poor")
```

```python
#problem 8
"""
Kiểm tra năm nhuận: 
Viết chương trình nhập năm, kiểm tra xem năm đó có phải năm nhuận hay không.  
(Năm nhuận nếu chia hết cho 400 hoặc chia hết cho 4 nhưng không chia hết cho 100)
"""

a = int(input("Enter the year: "))
if a % 400 == 0 or (a % 4 == 0 and a % 100 != 0):
    print("This is a leap year")
else:
    print("This is not a leap year")
```

```python
#problem 9
"""
Chuyển đổi số thành chữ số: 
Nhập một số nguyên từ 0 đến 9, in ra tên chữ số tương ứng (0 -> "Không", 1 -> "Một", ...)  
→ Gợi ý: dùng match … case
"""

a = int(input("Enter a number (0-10): "))
if a == 0:
    print("Zero")
elif a == 1:
    print("One")
elif a == 2:
    print("Two")
elif a == 3:
    print("Three")
elif a == 4:
    print("Four")
elif a == 5:
    print("Five")
elif a == 6:
    print("Six")
elif a == 7:
    print("Seven")
elif a == 8:
    print("Eight")
elif a == 9:
    print("Nine")
elif a == 10:
    print("Ten")
else:
    print("Out of range (0–10 only)")
```
