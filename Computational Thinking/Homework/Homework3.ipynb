```python
#Problem 4
h = int(input("Nhập chiều cao h: "))

for i in range(1, h + 1):
    space = " " * (h - i)
    middle = " " * (2 * i - 3) if i != h else "*" * (2 * i - 3)
    left = "*" if 2 * i - 3 > 0 else ""
    print(space + "*" + middle + left)
```

```python
#Problem 14
max_n = None
min_n = None
while True:
    n = int(input())
    if n == -1:
        break
    if max_n == None or n > max_n:
        max_n = n
    if min_n == None or n < min_n:
        min_n = n
print(max_n, min_n)
        
    
```

```python
#Problem 16
n = int(input())
chan = 0
le = 0
while n > 0:
    check = n % 10
    if check % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10
print(chan, le)
```

```python
#Problem 20
n = int(input())
s = 0
while n > 0:
    i = n % 10
    check = True
    if i < 2:
        check = False
    for j in range(2, int(i**0.5) + 1):
        if i % j == 0:
            check = False
    if check == True:
        s += 1
    n //= 10
print(s)
        
```
