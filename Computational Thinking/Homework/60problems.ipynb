```python
#Problem 50
a = list(map(int, input().split()))
b = list(map(int, input().split()))
def poly(p):
    s = []
    for i, c in enumerate(p):
        if c == 0:
            continue
        elif i == 0:
            s.append(str(c))
        elif c == 1:
            s.append("x" if i == 1 else f"x^{i}")
        else:
            s.append(f"{c}x" if i == 1 else f"{c}x^{i}")
    return " + ".join(s) if s else "0"
max_len = max(len(a), len(b))
a += [0]*(max_len - len(a))
b += [0]*(max_len - len(b))
total = [a[i] + b[i] for i in range(max_len)]
print(f"({poly(a)}) + ({poly(b)}) = {poly(total)}")
```

```python
#Problem 55
n = int(input())

# Nhập ma trận A
A = []
for i in range(n):
    row = list(map(int, input().split()))
    A.append(row)

# Nhập ma trận B
B = []
for i in range(n):
    row = list(map(int, input().split()))
    B.append(row)

# a) Tổng hai ma trận
C = [[0]*n for _ in range(n)]
for i in range(n):
    for j in range(n):
        C[i][j] = A[i][j] + B[i][j]

# b) Tích hai ma trận
D = [[0]*n for _ in range(n)]
for i in range(n):
    for j in range(n):
        for k in range(n):
            D[i][j] += A[i][k] * B[k][j]

# In kết quả
print("Tong hai ma tran:")
for row in C:
    print(*row)

print("Tich hai ma tran:")
for row in D:
    print(*row)
```

```python
#2.
a = list(map(int, input().split()))
le = 0
chan = 0
for i in a:
    if i % 2 == 0:
        chan += 1
    if i % 2 != 0:
        le += 1
if chan == 0:
    for i in range(len(a)):
        a.append(2)
elif le == 0:
    for i in range(len(a)):
        a.append(1)
elif chan == le:
    print("Ham can bang")
elif chan > le:
    s = 0
    for i in range(len(a)):
        if i % 2!= 0:
            s = a[0]
            if s < a[i]:
                s = a[i]
    for i in range(chan - le):
        a.append(s)
elif le > chan:
    t = 0
    for i in range(len(a)):
        if i % 2 == 0:
            t = a[0]
            if t < a[i]:
                t = a[i]
    for i in range(le - chan):
        a.append(t)
    
print(a)
            
```
