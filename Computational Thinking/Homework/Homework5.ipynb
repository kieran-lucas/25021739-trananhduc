```python
#W6A17
n = int(input())
A = [list(map(int, input().split())) for i in range(n)]
print("Duong cheo chinh")
for i in range(n):
    print(A[i][i], end = " ")
print("\nDuong cheo phu")
for i in range(n):
    print(A[i][n - 1 - i], end = " ")
```

```python
#W6A21
n = list(map(str, input().split()))
d = { }
for i, j in enumerate(n):
    if j not in d:
        d[j] = []
        d[j].append(i)
for k in d:
    d[k] = tuple(d[k])
print(d)
```

```python
#W6A26
n = list(map(int, input().split()))
s = 0
for i in range(1, len(n) - 1):
    if n[i] > n[i - 1] and n[i] > n[i + 1]:
        s += n[i]
print(s)
    
```

```python
#W6A27
n = list(map(int, input().split()))
x = int(input())
found = False
for i in range(len(n)):
    if n[i] < x:
        print(i, end = " ")
        found = True
if found == False:
    print(-1)
```

```python
#W6A28
n = list(map(int, input().split()))
inc = True
dec = True
for i in range(len(n) - 1):
    if n[i] >= n[i + 1]:
        inc = False
    if n[i] <= n[i + 1]:
        dec = False
if inc:
    print("Tang")
elif dec:
    print("Giam")
else:
    print("Vo danh")
```
